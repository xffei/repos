```python
import pandas as pd   

data = pd.read_csv('data\类别变量练习数据.csv', encoding='gbk')  
print(data.head())
```

```python
import matplotlib.pyplot as plt  
  
# 计算每个职业的频数  
father_occupations = data['父亲工作'].value_counts()  
mother_occupations = data['母亲工作'].value_counts()  
  
# 绘制饼图  
fig, ax = plt.subplots()  
ax.pie(father_occupations, labels=father_occupations.index, autopct='%1.1f%%')  
ax.pie(mother_occupations, labels=mother_occupations.index, autopct='%1.1f%%')  
plt.title('父母工作占比')  
plt.show()
```

```python
pivot_table = pd.pivot_table(data, index='父亲工作', columns='母亲工作', aggfunc='size')  
print(pivot_table)
```

```python
print(type(data['父亲工作']))
print(type(data['母亲工作']))
```

```python
# 将'父亲工作'和'母亲工作'列转换为数值型数据  
data['父亲工作_factorized'] =  pd.factorize(data['父亲工作'])[0]
data['母亲工作_factorized'] =  pd.factorize(data['母亲工作'])[0]
```

```python
from scipy.stats import pearsonr  
  
# 计算相关系数和p值  
corr, p = pearsonr(data['父亲工作_factorized'], data['母亲工作_factorized'])  
print("Pearson correlation coefficient:", corr)  
print("P-value:", p)
```
